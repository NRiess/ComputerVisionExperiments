## 1. Colour conversion 

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
img = cv.imread('img1.jpg')

# convert img into HSV space
img_hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)
cv.imshow('image in HSV space', img_hsv)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# convert img into HSV space
img_gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
cv.imshow('image in gray', img_gray)
cv.waitKey(0)
cv.destroyAllWindows()
```

## 2. Geometric transormation
### 2.1. Scaling

```python
height, width = img.shape[:2]
h_scale = 0.5
v_scale = 0.4
new_heigth = (int) (height*v_scale) 
new_width = (int) (width*h_scale) 
img_resize = cv.resize(img, (new_width, new_heigth), interpolation = cv.INTER_LINEAR)

cv.imshow('resize', img_resize)
cv.waitKey(0)
cv.destroyAllWindows()
```

### 2.2. Translation

```python
t_x = 100
t_y = 200
M = np.float32([[1,0,t_x],[0,1,t_y]])
height, width = img.shape[:2] 
img_translation = cv.warpAffine(img, M, (width, height))
cv.imshow('translation', img_translation)
cv.waitKey(0)
cv.destroyAllWindows()
```

### 2.3. Rotation

```python
theta = 45 
c_x = (width - 1) / 2.0 
c_y = (height - 1) /2.0 
c = (c_x, c_y) 
s = 1
M = cv.getRotationMatrix2D(c, theta, s)
height, width = img.shape[:2]
img_rotation = cv.warpAffine(img, M, (width, height))
cv.imshow('rotation', img_rotation)
cv.waitKey(0)
cv.destroyAllWindows()
```

### 2.4 Affine

```python
m00 = 0.38 
m01 = 0.27 
m02 = -47.18 
m10 = -0.14 
m11 = 0.75 
m12 = 564.32
M = np.float32([[m00, m01, m02], [m10, m11, m12]])
height, width = img.shape[:2]
img_affine = cv.warpAffine(img, M, (width, height))
cv.imshow('affine', img_affine)
cv.waitKey(0)
cv.destroyAllWindows()
```
